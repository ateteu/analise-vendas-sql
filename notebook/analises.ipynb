## Imports

```python
import pandas as pd
import sqlite3  
```

## Criar banco SQLite em memória

```python
produtos = pd.read_csv("../data/produtos.csv")
vendas   = pd.read_csv("../data/vendas.csv")

bd = sqlite3.connect(':memory:') # Criar BD na memória RAM
```

```python
produtos.to_sql(
    'produtos', 
    bd, 
    index=False, 
    if_exists='replace'
)

vendas.to_sql(
    'vendas',
    bd,
    index=False,
    if_exists='replace'
)
```
